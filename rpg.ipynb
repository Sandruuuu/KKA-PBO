```python
from abc import ABC, abstractmethod

# ============================================================
# LATIHAN 5: Abstract Base Class — Kontrak/Interface
# ============================================================
class GameUnit(ABC):
    @abstractmethod
    def serang(self, lawan=None):
        pass

    @abstractmethod
    def info(self):
        pass


# ============================================================
# LATIHAN 1, 2, 4: Class Hero dengan Enkapsulasi
# ============================================================
class Hero(GameUnit):
    def __init__(self, name, hp, attack_power):
        self.name = name                    # Nama Hero
        self.__hp = hp                      # Enkapsulasi: HP bersifat Private
        self.attack_power = attack_power    # Kekuatan Serangan

    # GETTER: Cara resmi melihat HP
    def get_hp(self):
        return self.__hp

    # SETTER: Cara resmi mengubah HP (dengan validasi)
    def set_hp(self, nilai_baru):
        if nilai_baru < 0:
            self.__hp = 0                   # HP tidak boleh negatif
        elif nilai_baru > 1000:
            print("Cheat terdeteksi! HP dimaksimalkan ke 1000 saja.")
            self.__hp = 1000
        else:
            self.__hp = nilai_baru

    def info(self):
        print(f"Hero: {self.name} | HP: {self.get_hp()} | Power: {self.attack_power}")

    def serang(self, lawan=None):
        if lawan:
            print(f"{self.name} menyerang {lawan.name}!")
            lawan.diserang(self.attack_power)
        else:
            print(f"{self.name} menyerang dengan tangan kosong.")

    def diserang(self, damage):
        sisa_hp = self.get_hp() - damage
        self.set_hp(sisa_hp)
        print(f"{self.name} terkena damage {damage}. Sisa HP: {self.get_hp()}")


# ============================================================
# LATIHAN 3: Inheritance — Class Mage
# ============================================================
class Mage(Hero):
    def __init__(self, name, hp, attack_power, mana):
        super().__init__(name, hp, attack_power)  # Memanggil constructor Parent
        self.mana = mana

    def info(self):
        print(f"{self.name} [Mage] | HP: {self.get_hp()} | Mana: {self.mana}")

    def serang(self, lawan=None):
        if lawan:
            print(f"{self.name} (Mage) menembakkan Bola Api ke {lawan.name}! Boom!")
            lawan.diserang(self.attack_power)
        else:
            print(f"{self.name} (Mage) menembakkan Bola Api! Boom!")

    def skill_fireball(self, lawan):
        if self.mana >= 20:
            print(f"{self.name} menggunakan Fireball ke {lawan.name}!")
            self.mana -= 20
            lawan.diserang(self.attack_power * 2)   # Damage 2x lipat
        else:
            print(f"{self.name} gagal skill! Mana tidak cukup.")


# ============================================================
# LATIHAN 6: Polymorphism — Berbagai tipe Hero
# ============================================================
class Archer(Hero):
    def info(self):
        print(f"{self.name} [Archer] | HP: {self.get_hp()} | Power: {self.attack_power}")

    def serang(self, lawan=None):
        print(f"{self.name} (Archer) memanah dari jauh! Jleb!")
        if lawan:
            lawan.diserang(self.attack_power)


class Fighter(Hero):
    def info(self):
        print(f"{self.name} [Fighter] | HP: {self.get_hp()} | Power: {self.attack_power}")

    def serang(self, lawan=None):
        print(f"{self.name} (Fighter) memukul dengan pedang! Slash!")
        if lawan:
            lawan.diserang(self.attack_power)


class Healer(Hero):
    def info(self):
        print(f"{self.name} [Healer] | HP: {self.get_hp()} | Power: {self.attack_power}")

    def serang(self, lawan=None):
        print(f"{self.name} tidak menyerang, tapi menyembuhkan teman!")


# ============================================================
# LATIHAN 5: Monster sebagai implementasi GameUnit
# ============================================================
class Monster(GameUnit):
    def __init__(self, jenis, hp):
        self.name = jenis
        self.__hp = hp

    def get_hp(self):
        return self.__hp

    def set_hp(self, nilai):
        self.__hp = max(0, nilai)

    def diserang(self, damage):
        self.set_hp(self.get_hp() - damage)
        print(f"{self.name} terkena damage {damage}. Sisa HP: {self.get_hp()}")

    def serang(self, lawan=None):
        target = lawan.name if lawan else "target"
        print(f"Monster {self.name} menggigit {target}!")
        if lawan:
            lawan.diserang(10)

    def info(self):
        print(f"Monster: {self.name} | HP: {self.get_hp()}")


print("✅ Semua class berhasil dimuat!")
```

```python
# ============================================================
# LATIHAN 1: Class & Object Dasar
# ============================================================
print("=" * 55)
print("LATIHAN 1: Class & Object Dasar")
print("=" * 55)

hero1 = Hero("Layla", 100, 15)
hero2 = Hero("Zilong", 120, 20)

print("\n--- Info Hero ---")
hero1.info()
hero2.info()

# Tugas Analisis 1: Mengubah HP setelah objek dibuat
hero1.set_hp(500)
print(f"\nSetelah hero1.set_hp(500) → HP Layla: {hero1.get_hp()}")


# ============================================================
# LATIHAN 2: Interaksi Antar Objek (Method)
# ============================================================
print("\n" + "=" * 55)
print("LATIHAN 2: Interaksi Antar Objek")
print("=" * 55)

hero1 = Hero("Layla", 100, 15)
hero2 = Hero("Zilong", 120, 20)

print("\n--- Pertarungan Dimulai ---")
hero1.serang(hero2)   # Layla menyerang Zilong
hero2.serang(hero1)   # Zilong membalas


# ============================================================
# LATIHAN 3: Pewarisan (Inheritance)
# ============================================================
print("\n" + "=" * 55)
print("LATIHAN 3: Pewarisan (Inheritance)")
print("=" * 55)

eudora = Mage("Eudora", 80, 30, 100)
balmond = Hero("Balmond", 200, 10)

print("\n--- Update Class Hero ---")
eudora.info()
eudora.serang(balmond)          # Serangan biasa (warisan dari Hero)
eudora.skill_fireball(balmond)  # Skill khusus Mage


# ============================================================
# LATIHAN 4: Enkapsulasi
# ============================================================
print("\n" + "=" * 55)
print("LATIHAN 4: Enkapsulasi")
print("=" * 55)

hero_test = Hero("Layla", 100, 15)

print("\n--- Uji Validasi Setter ---")
hero_test.set_hp(-50)
print(f"HP setelah set ke -50   : {hero_test.get_hp()}")

hero_test.set_hp(9999)
print(f"HP setelah set ke 9999  : {hero_test.get_hp()}")

# Name Mangling (Tugas Analisis 4)
print(f"\nAkses Name Mangling     : {hero_test._Hero__hp}")


# ============================================================
# LATIHAN 5: Abstraksi (Abstract Class)
# ============================================================
print("\n" + "=" * 55)
print("LATIHAN 5: Abstraksi (Abstract Class)")
print("=" * 55)

h = Hero("Alucard", 150, 25)
m = Monster("Serigala", 80)

print("\n--- Info Unit ---")
h.info()
m.info()

print("\n--- Pertarungan Hero vs Monster ---")
h.serang(m)
m.serang(h)


# ============================================================
# LATIHAN 6: Polymorphism
# ============================================================
print("\n" + "=" * 55)
print("LATIHAN 6: Polymorphism")
print("=" * 55)

pasukan = [
    Mage("Eudora",   80,  30, 100),
    Archer("Miya",   90,  25),
    Fighter("Zilong",150, 35),
    Mage("Gord",     75,  40, 150),
    Healer("Rafaela",100, 10),
]

print("\n--- PERANG DIMULAI ---")
for pahlawan in pasukan:
    pahlawan.serang()   # Satu perintah → respons berbeda (Polymorphism)
```

# **ANALISIS**

1. **Class, Object & Interaction**

    Enkapsulasi Dasar: Mengubah hero1.hp secara langsung tidak akan berhasil jika menggunakan atribut private (__hp). Perubahan harus melalui Setter agar validasi data terjaga.
    Interaksi Objek: Method serang(lawan) menerima objek utuh agar bisa memanggil method milik lawan (seperti lawan.diserang()), bukan sekadar teks nama.

2. **Inheritance (Pewarisan)**

    Pentingnya super(): Menghapus super().__init__() menyebabkan error karena atribut induk (nama, hp) tidak terinisialisasi. super() adalah jembatan wajib untuk mewariskan sifat induk ke anak.

3. **Encapsulation (Keamanan Data)**

    Name Mangling: Atribut private bisa diakses paksa lewat _Class__variable, tapi ini praktik buruk yang merusak struktur kode.
    Validasi: Setter berfungsi mencegah data tidak masuk akal (contoh: HP negatif) masuk ke dalam sistem.

4. **Abstraction (Kontrak Kode)**

    Abstract Class: Kelas abstrak (seperti GameUnit) tidak bisa dibuat objeknya karena hanya berupa "cetak biru". Ia memaksa kelas anak untuk mengimplementasikan method wajib agar program tidak crash.

5. **Polymorphism (Fleksibilitas)**

    Skalabilitas: Memungkinkan penambahan karakter baru (seperti Healer) tanpa mengubah logika looping utama.
    Konsistensi: Syarat utamanya adalah nama method harus sama (misal: semua pakai serang()). Jika nama berbeda, sistem otomatis gagal mengenali perintah tersebut.
